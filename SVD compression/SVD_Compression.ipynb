# Singular Value Decomposition Image Compression

```python
import numpy as np
def outer_product(u, s, v):
    n = u.shape[1] # number of columns in U array
    A = np.zeros((n,n,n))
    for i in range(n):
        A[i] = np.matrix(u[:, :i]) * np.diag(s[:i]) * np.matrix(v[:i, :])
    return A

def compress(img, size, n):
    output = np.zeros((size, size))
    for i in range(n):
        output += img[i]
    return output
```

```python
from matplotlib import pyplot as plt
import cv2

folder_path = './images/'
og_img = cv2.imread(folder_path + 'oldman.jpg')
gray_img = cv2.cvtColor(og_img, cv2.COLOR_BGR2GRAY)

size = min(og_img.shape[:2])
square_img = gray_img[0:size, 0:size]

img_arr = np.array(square_img)

plt.imshow(square_img)
plt.title('Biden')
plt.show()
```

```python
u, s, vh = np.linalg.svd(img_arr)
svd = outer_product(u, s, vh)

for n in range(1, 102, 25):
    plt.imshow(svd[n])
    plt.title('Old Man n = '+str(n))
    plt.show()
```
